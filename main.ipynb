# Tiền xử lý dữ liệu 

```python
# Kiểm tra và xử lý giá trị trùng lặp trong các cột chính
print("=" * 30)
print("1. KIỂM TRA DỮ LIỆU TRÙNG LẶP")


# Tìm bản ghi trùng lặp toàn bộ
duplicates = df.duplicated().sum()
print(f"Tổng bản ghi trùng lặp 100%: {duplicates}")

# Tìm trùng lặp dựa trên ID + Tên môn + Học kỳ (chỉ báo trùng lặp thực sự)
key_duplicates = df.duplicated(subset=['ID', 'Tên môn', 'Học kỳ'], keep=False).sum()
print(f"Trùng lặp trên ID + Tên môn + Học kỳ: {key_duplicates}")

# Loại bỏ bản ghi trùng lặp hoàn toàn (giữ lại bản ghi đầu tiên)
df_cleaned = df.drop_duplicates(keep='first')
print(f"Số bản ghi sau loại bỏ trùng lặp: {len(df_cleaned)}")
```

```python
# Chuẩn hóa cột "Tên lớp"
print("2. CHUẨN HÓA CỘT TÊN LỚP")


# Loại bỏ khoảng trắng thừa
df_cleaned['Tên lớp'] = df_cleaned['Tên lớp'].str.strip()
df_cleaned['Tên lớp.1'] = df_cleaned['Tên lớp.1'].str.strip()

print(f"Các giá trị Tên lớp sau chuẩn hóa:")
print(df_cleaned['Tên lớp'].value_counts().head(10))
```

```python
# Xử lý cột "Học kỳ"
print("3. CHUẨN HÓA CỘT HỌC KỲ")


# Tách học kỳ thành kỳ chính và kỳ phụ
df_cleaned['Học kỳ'] = df_cleaned['Học kỳ'].str.strip()
df_cleaned['Là_học_kỳ_phụ'] = df_cleaned['Học kỳ'].str.contains('phu', case=False, na=False).astype(int)

# Tạo cột học kỳ chuẩn (chỉ lấy số)
df_cleaned['Học kỳ_chuẩn'] = df_cleaned['Học kỳ'].str.extract(r'(\d+)', expand=False)

print(f"Các giá trị Học kỳ duy nhất: {df_cleaned['Học kỳ'].unique()}")
print(f"\nPhân bố học kỳ phụ:")
print(df_cleaned['Là_học_kỳ_phụ'].value_counts())
```

```python
#  Xử lý cột "Địa chỉ thường trú"
print("4. CHUẨN HÓA ĐỊA CHỈ")

# Loại bỏ khoảng trắng, chuyển thành hoa
df_cleaned['Địa chỉ thường trú'] = df_cleaned['Địa chỉ thường trú'].str.strip().str.upper()

# Xác định tỉnh từ địa chỉ (lấy từ cuối cùng)
df_cleaned['Tỉnh_thành'] = df_cleaned['Địa chỉ thường trú'].str.split(',').str[-1].str.strip()

print(f"Top 15 Tỉnh/Thành:")
print(df_cleaned['Tỉnh_thành'].value_counts().head(15))
```

```python
#  Xử lý cột "Trạng thái"
print("5. CHUẨN HÓA CỘT TRẠNG THÁI")

df_cleaned['Trạng thái'] = df_cleaned['Trạng thái'].str.strip()
print(f"Các giá trị Trạng thái:")
print(df_cleaned['Trạng thái'].value_counts())
```

```python
#  Xử lý cột điểm (chuyển sang numeric, xử lý NaN)
print("6. CHUẨN HÓA CÁC CỘT ĐIỂM")
score_cols = ['Điểm chuyên cần 1', 'Điểm trung bình thực hành', 'Điểm trung bình thường kỳ', 
              'Điểm thi', 'Điểm tổng kết', 'Điểm tín chỉ']

for col in score_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
        missing = df_cleaned[col].isna().sum()
        print(f"{col}: {missing} giá trị NaN, Min: {df_cleaned[col].min():.2f}, Max: {df_cleaned[col].max():.2f}")
```

```python
# Xử lý "Số buổi nghỉ" - kiểm tra giá trị âm hoặc quá lớn
print("KIỂM TRA CỘT SỐ BUỔI NGHỈ")
df_cleaned['Số buổi nghỉ'] = pd.to_numeric(df_cleaned['Số buổi nghỉ'], errors='coerce')
anomaly_attendance = df_cleaned[(df_cleaned['Số buổi nghỉ'] < 0) | (df_cleaned['Số buổi nghỉ'] > 100)]
print(f"Giá trị bất thường trong 'Số buổi nghỉ': {len(anomaly_attendance)}")
if len(anomaly_attendance) > 0:
    print(anomaly_attendance[['ID', 'Tên lớp', 'Tên môn', 'Số buổi nghỉ']])
```

```python
#  Xử lý cột "Vắng thi"
print(" KIỂM TRA CỘT VẮNG THI")
df_cleaned['Vắng thi'] = pd.to_numeric(df_cleaned['Vắng thi'], errors='coerce')
print(f"Bản ghi vắng thi > 0: {(df_cleaned['Vắng thi'] > 0).sum()}")
print(f"Phân bố vắng thi:")
print(df_cleaned['Vắng thi'].value_counts().head())
```

```python
#  Xử lý cột "Không tính trung bình tích lũy"
print("9. KIỂM TRA CỘT KHÔNG TÍNH GPA")
df_cleaned['Không tính trung bình tích lũy'] = df_cleaned['Không tính trung bình tích lũy'].fillna('').str.strip()
print(f"Giá trị trong cột 'Không tính trung bình tích lũy':")
print(df_cleaned['Không tính trung bình tích lũy'].value_counts())
```

```python
#  Xử lý "Loại môn học"
print("10. KIỂM TRA LOẠI MÔN HỌC")
df_cleaned['Loại môn học'] = df_cleaned['Loại môn học'].str.strip()
print(f"Phân bố loại môn học:")
print(df_cleaned['Loại môn học'].value_counts())
```

```python
# Xử lý cột "Điểm chuyên cần 2" (cột dư thừa)
print("11. KIỂM TRA CỘT ĐIỂM CHUYÊN CẦN 2 (DƯ THỪA)")


non_null_attendance2 = df_cleaned['Điểm chuyên cần 2'].notna().sum()
print(f"Số bản ghi có giá trị trong 'Điểm chuyên cần 2': {non_null_attendance2}")
if non_null_attendance2 < 3:
    print("=> Nên xóa cột này để tránh nhiễu dữ liệu")
    df_cleaned = df_cleaned.drop(columns=['Điểm chuyên cần 2'])
```

```python
# 12. Kiểm tra và xử lý ngoại lệ điểm
print("12. KIỂM TRA NGOẠI LỆ ĐIỂM")
# Điểm thi = 0 hoặc = 10
zero_exams = (df_cleaned['Điểm thi'] == 0).sum()
perfect_exams = (df_cleaned['Điểm thi'] == 10).sum()
print(f"Điểm thi = 0: {zero_exams}")
print(f"Điểm thi = 10: {perfect_exams}")

# Điểm tổng kết không nằm trong [0, 10]
out_of_range = ((df_cleaned['Điểm tổng kết'] < 0) | (df_cleaned['Điểm tổng kết'] > 10)).sum()
print(f"Điểm tổng kết ngoài phạm vi [0, 10]: {out_of_range}")
```

```python

# 13. Kiểm tra tính nhất quán điểm tín chỉ
print("13. KIỂM TRA NHẤT QUÁN ĐIỂM TÍN CHỈ")

# Công thức: Điểm tín chỉ = Điểm tổng kết * Số tín chỉ / 10
valid_credit_rows = (df_cleaned['Điểm tín chỉ'].notna()) & (df_cleaned['Điểm tổng kết'].notna()) & (df_cleaned['Số tín chỉ'].notna())
if valid_credit_rows.sum() > 0:
    expected_credit = (df_cleaned.loc[valid_credit_rows, 'Điểm tổng kết'] * 
                      df_cleaned.loc[valid_credit_rows, 'Số tín chỉ'] / 10).round(2)
    actual_credit = df_cleaned.loc[valid_credit_rows, 'Điểm tín chỉ'].round(2)
    inconsistent = (expected_credit != actual_credit).sum()
    print(f"Bản ghi điểm tín chỉ không khớp công thức: {inconsistent}/{valid_credit_rows.sum()}")
```

```python
# Chuẩn hóa các cột dạng text: loại bỏ khoảng trắng thừa, chuyển về định dạng thống nhất
# DO Khoảng trắng thừa gây sai lệch khi phân nhóm, so sánh hoặc thống kê.
df['Tên lớp'] = df['Tên lớp'].str.strip()
df['Tên lớp.1'] = df['Tên lớp.1'].str.strip()
df['Học kỳ'] = df['Học kỳ'].str.strip()
df['Trạng thái'] = df['Trạng thái'].str.strip()
df['Địa chỉ thường trú'] = df['Địa chỉ thường trú'].str.strip().str.upper()
df['Loại môn học'] = df['Loại môn học'].str.strip()
```

```python
# Tách học kỳ phụ và chuẩn hóa học kỳ
# DO Học kỳ phụ có thể ảnh hưởng đến tiến độ học tập, cần phân biệt rõ để phân tích chính xác.
df['Là_học_kỳ_phụ'] = df['Học kỳ'].str.contains('phu', case=False, na=False).astype(int)
df['Học kỳ_chuẩn'] = df['Học kỳ'].str.extract(r'(\d+)', expand=False)
```

```python
#  Chuẩn hóa địa chỉ thường trú, tách tỉnh/thành
# DO Địa chỉ chuẩn hóa giúp phân tích vùng miền, xác định khu vực có tỷ lệ thôi học cao hoặc điểm thấp.
df['Tỉnh_thành'] = df['Địa chỉ thường trú'].str.split(',').str[-1].str.strip()
```

```python
# Chuẩn hóa các cột điểm về kiểu số, xử lý giá trị NaN
# DO Đảm bảo các phép tính toán, thống kê không bị lỗi do dữ liệu kiểu text hoặc thiếu giá trị.
score_cols = ['Điểm chuyên cần 1', 'Điểm trung bình thực hành', 'Điểm trung bình thường kỳ', 
              'Điểm thi', 'Điểm tổng kết', 'Điểm tín chỉ']
for col in score_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```
